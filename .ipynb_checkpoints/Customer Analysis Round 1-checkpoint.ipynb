```python
import pandas as pd
import numpy as np
```

Read the three files into python as dataframes

```python
file1 = pd.read_csv('./file1.csv')
file2 = pd.read_csv('./file2.csv')
file3 = pd.read_csv('./file3.csv')
```

Show the DataFrame's shape

```python
display(file1.shape)
display(file2.shape)
display(file3.shape)
```

Standardize header names

```python
display(file1.columns)
display(file2.columns)
display(file3.columns)
```

```python
file1 = file1.rename(columns={'ST':'State', 'GENDER':'Gender'})
file2 = file2.rename(columns={'ST':'State', 'GENDER':'Gender'})
#I have used file3 as the template
display(file1.columns)
display(file2.columns)
display(file3.columns)
```

Rearrange the columns in the dataframe as needed

```python
file1 = file1[['Customer', 'State', 'Customer Lifetime Value', 'Education', 'Gender',
       'Income', 'Monthly Premium Auto', 'Number of Open Complaints',
       'Policy Type', 'Total Claim Amount', 'Vehicle Class']]
file2 = file2[['Customer', 'State', 'Customer Lifetime Value', 'Education', 'Gender',
       'Income', 'Monthly Premium Auto', 'Number of Open Complaints',
       'Policy Type', 'Total Claim Amount', 'Vehicle Class']]
#I have used file3 as the template
display(file1.columns)
display(file2.columns)
display(file3.columns)
```

Concatenate the three dataframes

```python
data = pd.concat([file1, file2, file3], axis=0)
data
```

```python
data = data.reset_index()
data = data.drop(['index'], axis=1)
data
```

Correct the values in the column customer lifetime value. They are given as a percent, so multiply them by 100 and change dtype to numerical type

```python
data['Customer Lifetime Value']
```

As the values on the first two files had a % symbol on them but the third did not I chose to clean them before merging

```python
file1["Customer Lifetime Value"] = file1["Customer Lifetime Value"].str.replace("%","")
file2["Customer Lifetime Value"] = file2["Customer Lifetime Value"].str.replace("%","")
clean_data = pd.concat([file1, file2, file3], axis=0)
```

I have taken out the % symbol but the number is still too big, I may have to fix it later

```python
clean_data
```

```python
clean_data['Customer Lifetime Value'] =  pd.to_numeric(clean_data['Customer Lifetime Value'], errors='raise', downcast='float')
clean_data.dtypes
```

```python
clean_data
```

Now the column has a numerical type and I can multiply it by 100 (even if the numbers seem to big)

```python
clean_data['Customer Lifetime Value'] = clean_data['Customer Lifetime Value']/100
clean_data
```

Check for duplicate rows in the data and remove if any

```python
clean_data = clean_data.drop_duplicates()
clean_data.shape
```
